```python
list = []
addr = []
gu = []
gu_count = []

with open('부산광역시_노인복지관 현황_20230101.csv', encoding='cp949') as f :       #cp949 : 코드페이지 949 인코딩 방식도 존재
    for line in f :
        list.append(line.split(','))
    
    for i in range(len(list)) :
        addr.append(list[i][2].split(' '))
    
    for i in range(1, len(addr)) :
        gu.append(addr[i][1])
    
    for i in set(gu) :
        gu_count.append(gu.count(i))
        print(f"{i} : {gu.count(i)}")
```
