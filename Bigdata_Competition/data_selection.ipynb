```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('full_data_add_price.csv', encoding='cp949')
df = df.iloc[:-1,1:]
```

```python
df.head()
```

```python
df.info()
```

```python
df_economy = pd.read_csv('고용지표_20240608223610.csv')
```

```python
df_economy.head()
```

```python
df_economy = df_economy.iloc[1:,[0,1,4,7]]

df_economy.head()
```

```python
df_economy['년'] = df_economy['시점'].str.split('.').str[0]
df_economy['월'] = df_economy['시점'].str.split('.').str[1]

df_economy['년'] = df_economy['년'].astype('int64')
df_economy['월'] = df_economy['월'].astype('int64')

df_economy.head()
```

```python
merged_df = pd.merge(df, df_economy, on=['년','월'], how='left')

merged_df.head()
```

```python
merged_df = merged_df.drop(columns='시점')

merged_df.head()
```

```python
merged_df.tail()
```

```python
merged_df.to_csv('full_data_add_economy.csv', encoding='cp949')
```
