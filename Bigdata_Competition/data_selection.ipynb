```python
df_economy = pd.read_csv('고용지표_20240608223610.csv')
```

```python
df_economy = df_economy.iloc[1:,[0,1,4,7]]

df_economy.head()
```

```python
df_economy['년'] = df_economy['시점'].str.split('.').str[0]
df_economy['월'] = df_economy['시점'].str.split('.').str[1]

df_economy['년'] = df_economy['년'].astype('int64')
df_economy['월'] = df_economy['월'].astype('int64')

df_economy.head()
```

```python
merged_df = pd.merge(df, df_economy, on=['년','월'], how='left')

merged_df.head()
```

```python
df_indicator = pd.read_excel('주요지표_분기지표__20240608225025.xlsx')

df_indicator.head()
```

```python
df_indicator_T = df_indicator.transpose()

df_indicator_T.head()
```

```python
df_indicator_T = df_indicator_T.iloc[:,:9]

df_indicator_T.head()
```

```python
df_indicator_T.reset_index(inplace=True)
df_indicator_T.columns = df_indicator_T.iloc[0]
df_indicator_T = df_indicator_T.iloc[1:]

df_indicator_T['년'] = df_indicator_T['계정항목별'].str.split('.').str[0]
df_indicator_T['분기'] = df_indicator_T['계정항목별'].str.split('.').str[1]

df_indicator_T['년'] = df_indicator_T['년'].astype('int64')

df_indicator_T.head()
```

```python
def get_quarter(month):
    if month in [1,2,3]:
        return 'Q1'
    elif month in [4,5,6]:
        return 'Q2'
    elif month in [7,8,9]:
        return 'Q3'
    elif month in [10,11,12]:
        return 'Q4'

merged_df['분기'] = merged_df['월'].apply(get_quarter)

merged_df.head()
    
```

```python
def convert_quarter(quarter):
    mapping = {'1/4': 'Q1', '2/4': 'Q2', '3/4': 'Q3', '4/4': 'Q4'}
    return mapping.get(quarter)

df_indicator_T['분기'] = df_indicator_T['분기'].apply(convert_quarter)

df_indicator_T.head()
```

```python
final_data = pd.merge(merged_df, df_indicator_T, on=['년','분기'], how='left')
```

```python
final_data = final_data.drop(columns=['계정항목별'])
final_data.head()
```
