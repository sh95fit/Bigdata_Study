```python
import pandas as pd
import numpy as np

from datetime import datetime
```

```python
w_data = pd.read_csv('total_weather.csv', encoding='cp949')

w_data = w_data.fillna(0)

# 년, 월, 일, 시간 분리
w_data['일시'] = pd.to_datetime(w_data['일시'], format='%Y-%m-%d %H:%M', errors='raise')
w_data['년'] = w_data['일시'].dt.year
w_data['월'] = w_data['일시'].dt.month
w_data['일'] = w_data['일시'].dt.day
w_data['시간'] = w_data['일시'].dt.hour
w_data['분'] = w_data['일시'].dt.minute

w_data.info()
```

```python
w_data.head()
```

```python
w_data.set_index('일시', inplace=True)
```

```python
w_data_15T = w_data.resample('15min').asfreq()

print(len(w_data_15T), len(w_data)*4)
```

```python
print(w_data_15T.head())
print(w_data_15T.tail())
```

```python
w_data_15T = w_data_15T.iloc[:-1,:]

w_data_15T.tail()
```

```python
w_data_15T['기온(°C)'] = w_data['기온(°C)'].resample('15min').interpolate(method='linear')
```

#### asfreq 대신 값을 모두 채워놓는 ffill을 활용해서 모두 채운 후 보간법 및 나누기 적용

```python
wdata_15T = w_data.resample('15min').ffill()
```

```python
wdata_15T.head()
```

```python
wdata_15T['기온(°C)'] = w_data['기온(°C)'].resample('15min').interpolate(method='linear')
wdata_15T['지면온도(°C)'] = w_data['지면온도(°C)'].resample('15min').interpolate(method='linear')
wdata_15T['습도(%)'] = w_data['습도(%)'].resample('15min').interpolate(method='linear')
wdata_15T['현지기압(hPa)'] = w_data['현지기압(hPa)'].resample('15min').interpolate(method='linear')
wdata_15T['일사(MJ/m2)'] = w_data['일사(MJ/m2)'].resample('15min').interpolate(method='linear')
```

```python
wdata_15T.head()
```

```python
for col in ['강수량(mm)', '일조(hr)', '적설(cm)']:
    wdata_15T[col] = w_data[col].resample('15min').ffill() / 4
```

```python
wdata_15T = wdata_15T.fillna(0)
```

```python
wdata_15T.to_csv('final_weather.csv', encoding='cp949')
```
