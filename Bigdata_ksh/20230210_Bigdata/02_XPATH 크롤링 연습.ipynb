```python
# 네이버 별자리 운세
import requests
from lxml import etree

url = 'https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&qvt=0&query=%EB%B3%84%EC%9E%90%EB%A6%AC%20%EC%9A%B4%EC%84%B8'
p = requests.get(url)
```

```python
p.text
```

```python
p2 = etree.fromstring(p.text, parser=etree.HTMLParser())
```

```python
r = p2.xpath('//li[@class="lst_r"]/p')
print(r)
for i, v in enumerate(r) :
    print(i, v.text)
r = p2.xpath('//*[@id="main_pack"]/section[1]/div/div[2]/ul[2]/li[1]/p')
print("")
print(r)
print(r[0].text)
```
