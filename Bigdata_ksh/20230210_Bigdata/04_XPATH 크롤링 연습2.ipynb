```python
# 네이버 별자리 운세
import requests
from lxml import etree

url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver'
p = requests.get(url)
```

```python
p.text
```

```python
p2 = etree.fromstring(p.text, parser=etree.HTMLParser())
```

```python
r = p2.xpath('//*[@id="old_content"]/table/tbody/tr[2]/td[@class="title"]/div/a')
print(r)
print(r[0].text)
print('')
r = p2.xpath('//div[@class="tit3"]/a')
for i in r :
    print(i.text)
```

```python
for i in r :
    print(i.attrib)
    print(i.attrib['href'])
```

```python
sample_detail_page = r[0].attrib['href']
print(sample_detail_page)
```

```python
new_page = requests.get("https://movie.naver.com"+sample_detail_page)
```

```python
new_page.text
```

```python
p2 = etree.fromstring(new_page.text, parser=etree.HTMLParser())
```

```python
r = p2.xpath('//*[@id="actualPointPersentBasic"]/div/span/span')
r = p2.xpath('//div[@class="star_score"]/em')
# print(r)
score_string = ''
for i in r[:4]:
    # print(i.text)
    score_string = score_string + i.text
score_string = float(score_string)
print(score_string)
print(type(score_string))
```
