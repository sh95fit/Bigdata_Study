* 현재 상영영화 페이지에서 아래 데이터 수집, 리스트로 저장
 - 영화 제목
 - 평점(네티즌)
 - 추가 보너스 : 개봉일자, 상영시간

[['라스트버스, 9.60], ['원웨이', 5.75], ...]

```python
url = "https://movie.naver.com/movie/running/current.naver"

p = requests.get(url)
```

```python
p2 = etree.fromstring(p.text, etree.HTMLParser())
```

```python
title_list = []

title = p2.xpath('//*[@class="tit"]/a')

for i in title :
    title_list.append(i.text)

print(title_list)
```

```python
score_list = []

score = p2.xpath('//div[@class="star_t1"]/a/span[@class="num"]')
# score = p2.xpath('//dt[@class="tit_t1"]/following-sibling::*')
score = p2.xpath('//dt[@class="tit_t1"]/following-sibling::*/div[@class="star_t1"]/a/span[@class="num"]')
score = p2.xpath('//dt[@class="tit_t1"]/following::*/div[@class="star_t1"]/a/span[@class="num"]')

for i in score :
    score_list.append(i.text)

print(score_list)
```
