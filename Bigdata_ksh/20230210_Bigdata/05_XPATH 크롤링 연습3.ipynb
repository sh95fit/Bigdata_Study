* 현재 상영영화 페이지에서 아래 데이터 수집, 리스트로 저장
 - 영화 제목
 - 평점(네티즌)
 - 추가 보너스 : 개봉일자, 상영시간

[['라스트버스, 9.60], ['원웨이', 5.75], ...]

```python
url = "https://movie.naver.com/movie/running/current.naver"

p = requests.get(url)
```

```python
p2 = etree.fromstring(p.text, etree.HTMLParser())
```

```python
title_list = []

title = p2.xpath('//*[@class="tit"]/a')

for i in title :
    title_list.append(i.text)

print(title_list)
```

```python

score_list = []

score = p2.xpath('//div[@class="star_t1"]/a/span[@class="num"]')
# score = p2.xpath('//dt[@class="tit_t1"]/following-sibling::*/div[@class="star_t1"]/a/span[@class="num"]')
score = p2.xpath('//dt[@class="tit_t1"]/following::dd[1]/div[@class="star_t1"]/a/span[@class="num"]')
# score = p2.xpath('//a[contains(@href,"pointAfter")]/span[@class="num"]')
# score = p2.xpath('//dt[@class="tit_t1"]/following-sibling::dd[1]//span[@class="num"]')

for i in score :
    score_list.append(i.text)

print(score_list)
```

```python
print(len(title_list))
print(len(score_list))
```

```python
opening_date = []
running_time = []
total=[]

ods = p2.xpath('//dt[@class="tit_t1"]/following::dd[1]/span/following-sibling::text()')

for od in ods :
    # print(od.strip())
    total.append(od.strip())

while '' in total :
    total.remove('')

# print(total)

for t in range(0, len(total), 2) :
    running_time.append(total[t])
    opening_date.append(total[t+1])

print(opening_date)
print(running_time)
```

```python
print(len(opening_date))
print(len(running_time))
```

```python
total_list = list(zip(title_list, score_list, opening_date, running_time))
print(total_list)
```

```python
for i in range(len(total_list)) :
    print(f"{total_list[i][0]}\n평점 : {total_list[i][1]}\n상영시간 : {total_list[i][3]}\n개봉일 : {total_list[i][2]}\n")
```

```python
total_dict = {total_list[i][0] : list(total_list[i][1:]) for i in range(0,len(total_list),4)}
print(total_dict)
```
