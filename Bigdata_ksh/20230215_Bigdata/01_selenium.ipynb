```python
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
# 같은 경로에 위치시킨 경우
# driver = webdriver.Chrome()
# 다른 경로에 위치시킨 경우
driver = webdriver.Chrome('C:/Bigdata_ksh/20230215_Bigdata/chromedriver.exe')

url = 'https://movie.naver.com/movie/running/current.naver'

driver.get(url)
```

```python
r = driver.find_elements(By.XPATH, '//dt[@class="tit"]/a')
for i in r :
    print(i.text)
```

```python
r = driver.find_elements(By.XPATH, '//a[contains(@href,"pointAfter")]/span[@class="num"]')
for i in r :
    print(i.text)
```
