```python
import numpy as np
from matplotlib import image
import matplotlib.pyplot as plt
```

```python
cat_image = image.imread('cat.jpg')
print(cat_image.shape)
print(type(cat_image))

_, axe = plt.subplots()
axe.imshow(cat_image)
```

```python
cat_image_ch0 = cat_image[:,:,0]
print(cat_image_ch0.shape)

_, axe = plt.subplots()
axe.imshow(cat_image_ch0)
```

```python
cat_image_ch1 = cat_image[:,:,1]
print(cat_image_ch1.shape)

_, axe = plt.subplots()
axe.imshow(cat_image_ch1)
```

```python
cat_image_ch2 = cat_image[:,:,2]
print(cat_image_ch2.shape)

_, axe = plt.subplots()
axe.imshow(cat_image_ch2)
```

```python
cat_image_move_axis = np.moveaxis(cat_image, 2, 0)
print(cat_image_move_axis.shape)

_, axe = plt.subplots()
axe.imshow(cat_image_move_axis[2])
```
