```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd_data = pd.read_csv("전국평생학습강좌표준데이터.csv", encoding="cp949", header=None)
np_data = pd_data.to_numpy()
```

```python
print(np_data.shape)
for e in enumerate(np_data[0]):
    print(e)
```

```python
# 교육방법 구분 count (온라인 vs 그외)
filter1 = np_data[:,8] == "온라인"
# print(filter1)

a1 = np_data[filter1, :]
print("온라인 강좌 수 :", a1.shape[0])

filter2 = np_data[:,8] != "온라인"
# print(filter2)

a2 = np_data[filter2, :]
print("오프라인 강좌 수 :", a2.shape[0])

print("총 강좌 수 :", a1.shape[0]+a2.shape[0])
```

```python
# 교육방법 구분 count (모든 케이스 구분)
sub_data = np_data[1:, 8]
print(sub_data)
set1 = set(sub_data)
print(set1)

for n in set1 :
    filter1 = sub_data == n
    print(n, len(sub_data[filter1]))
```

```python
val, cnt = np.unique(sub_data, return_counts=True)
sub_dict = dict(zip(val, cnt))
print(sub_dict)
```

```python
# 직업능력개발훈련비지원강좌여부 == "Y"
sub_data_1 = np_data[1:, 21]
catagory, data_num = np.unique(sub_data_1, return_counts = True)
sub_dict_1 = dict(zip(catagory, data_num))
print(sub_dict_1)
```

```python
# 홈페이지 주소가 없는 것 count (아예 공백인 케이스만 산출 / 없음은 미포함)
filter2 = np_data[:, 20] == " "
a = np_data[filter2, :]
print(a.shape[1])
```

```python
# 운영기관별 강좌 개수
sub_data_2 = np_data[1:, 14]
name, cnt = np.unique(sub_data_2, return_counts = True)
sub_data_2 = dict(zip(name, cnt))
print(sub_data_2)
print(len(sub_data_2))
```
