```python
import numpy as np
import pandas as pd
```

```python
# 등록 인구 추이

np.set_printoptions(precision=3, suppress=True)

pd_data = pd.read_csv('등록인구추이_20230306201308.csv', encoding="cp949", header=None)
np_data = pd_data.to_numpy()
print(np_data)
print(np_data.shape)
```

```python
# 세대수 : 10만 단위 / 인구 : 100만 단위 / 한국인 : 100만 단위 / 외국인 : 100만 단위 /65세 이상 : 1000단위
a1 = np_data[1:,:]
# print(a1, a1.shape)
a1 = a1.astype(float)
# print(a1, a1.shape)
basis = np.array([1,100000,1000000,1000000,1000000,1000])
basis = basis.astype(float)
# print(basis, basis.shape)

sub_result = a1/basis

print(sub_result)
```

```python
# 각 년도별 : 한국인/인구 비율, 외국인/인구 비율
a2 = (sub_result[:, 3]/sub_result[:, 2])   # 한국인 / 인구 
a3 = (sub_result[:, 4]/sub_result[:, 2])   # 외국인 / 인구

print(sub_result[:,0])
print(a2)
print(a3)
```

```python
import matplotlib.pyplot as plt

_, axe = plt.subplots()
axe.plot(sub_result[:, 3])  # 한국인
axe.plot(sub_result[:, 4]*100)  # 외국인
```
