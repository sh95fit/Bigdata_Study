```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
```

```python
def fn1(ndarr) :
    return 5*ndarr+10

_, axe = plt.subplots()

xs = np.linspace(0, 100, 1000)
ys = fn1(xs)

axe.plot(xs, ys)
```

```python
# np.sqrt() : 제곱근 / np.pi : 파이 / np.exp : 지수

# Normal Distribution
def normal_distribution(x, sigma=1, mean=0) :
    return 1/np.sqrt(2*np.pi*sigma**2)*np.exp(-0.5*(((x-mean)/sigma)**2))


_, axe = plt.subplots()

xs = np.linspace(-5, 5, 100)
ys = normal_distribution(xs)

axe.plot(xs, ys)

ys1 = normal_distribution(xs, 2, 0)
axe.plot(xs,ys1)

ys2 = normal_distribution(xs, 1, 1)
axe.plot(xs,ys2)

ys3 = normal_distribution(xs, 1, -1)
axe.plot(xs,ys3)
```

```python
# CDF
import math
# math.erf() 활용

def normal_CDF(x, sigma=1, mean=0) :
    return 0.5*(1+scipy.special.erf((x-mean)/(sigma*np.sqrt(2))))

_, axe = plt.subplots()

xs = np.linspace(-5, 5, 100)

# 파랑
ys = normal_CDF(xs)
axe.plot(xs,ys)

# 노랑
ys1 = normal_CDF(xs, sigma=2, mean=0)
axe.plot(xs,ys1)

# 초록
ys2 = normal_CDF(xs, sigma=3, mean=0)
axe.plot(xs,ys2)

# 빨강
ys3 = normal_CDF(xs, sigma=1, mean=1)
axe.plot(xs,ys3)

# 보라
ys4 = normal_CDF(xs, sigma=1, mean=-1)
axe.plot(xs,ys4)
```
