```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

xs = np.linspace(-3, 3, 100)
ys = stats.norm.pdf(xs)     # pdf : 확률분포

_,axe = plt.subplots()

axe.plot(xs,ys)
```

```python
def count_with_replacement_order(pool, sel_count):  # all
    if sel_count == 1:
        return len(pool), [[x] for x in pool]
    a, b = count_with_replacement_order(pool, sel_count-1)
    r = []
    for x in pool:
        for b_i in b:
            r.append(b_i+[x])
    return len(r), r 
```

```python
# 앞 0 , 뒤 1 일 때, 동전을 10번 던져 나올 수  있는 합의 확률 분포
total_count, counts = count_with_replacement_order([0, 1], 10)
print(total_count)
print(counts)
count_np_1 = np.array(counts)
pass_1 = np.sum(count_np_1, axis=1)
pass_2 = np.unique(pass_1, return_counts=True)

_, axe = plt.subplots()
axe.plot(pass_2[0], pass_2[1]/total_count)

t_mean = np.mean(pass_1)
t_std = np.std(pass_1)
print(t_mean, t_std)

xs = np.linspace(0, 10, 100)
ys = stats.norm.pdf(xs, scale=t_std, loc=t_mean)

axe.plot(xs,ys)
```

```python
# 주사위를 5번 던졌을 때 
# 합이 25가 될 확률 : 0.016203703703703703
# 합이 25보다 크면 100원을 받기로 한다
# 합이 25보다 클 확률은? (25 불포함)

total_count, counts = count_with_replacement_order([x for x in range(1,7)], 5)
print(total_count)
print(counts)

count_np = np.array(counts)

pass_1 = np.sum(count_np, axis=1)
pass_2 = np.unique(pass_1, return_counts=True)
val, cnt = np.unique(pass_1, return_counts=True)

print(pass_2[0])
print(pass_2[1])

pass_3 = dict(zip(val,cnt))

print("합이 25가 될 확률 : ", pass_3[25]/total_count)
print("합이 25보다 클 확률 : ", np.sum(pass_2[1][-5:])/total_count)
```
