```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.api as sm

np.set_printoptions(precision=5, suppress=True)
```

```python
# normal curve
xs = np.linspace(-3, 3, 100)
ys = stats.norm.pdf(xs)

bot = xs[1]-xs[0]
print(np.sum(ys*bot))

_, axe = plt.subplots()

axe.plot(xs, ys)
```

```python
ys_cdf = stats.norm.cdf(xs)   # cumulative density function
_,axe = plt.subplots()
axe.plot(xs, ys_cdf)  # y값이 0.975일 때,  x r값 -> (거의) 1.96이 되어야 맞다
```

```python
# 밑변*높이의 각 요소
ys*bot

# 요소의 누적합
np.cumsum(ys*bot)

# 0.975에서 가장 가까운 값 찾기
a = (np.cumsum(ys*bot) - 0.975)**2
b = np.argmin(a)
print(b, xs[b])
```
