```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

fn = '전국평생학습강좌표준데이터.csv'

df = pd.read_csv(fn, encoding='cp949')

print(df.head())
```

```python
# columns의 인덱스 값, 이름 확인
for i, n in enumerate(df.columns) :
  print(i, n)
```

```python
# columns별 hasnans, dtype 확인
def df_check(df) :
  for i, col_name in enumerate(df.columns) :
    print(i, col_name, '\n', df[col_name].hasnans, df[col_name].dtype)

df_check(df)
```

#### 지자체별 온라인/오프라인 강좌 개수

```python
# 홈페이지 주소가 없는 강좌
homepage = df['홈페이지주소']

# print(homepage)

# print(homepage.hasnans)

print(homepage.isna().value_counts())
# print(homepage.unique())
# print(homepage.value_counts())

def myfn1(x) :
  try :
    return x[:5]
  except :
    return x

print(homepage.apply(myfn1))

http_types = homepage.apply(myfn1)
http_types.value_counts()
```

```python
df['교육방법구분'].unique()

edu_type = df['교육방법구분'].value_counts()

print(edu_type)
```

```python
## groupby 활용 시 간편

# 지자체별(시도별) 강좌 갯수

def myfn2(x) :
    try :
        return x.split(' ')[0]
    except :
        return x

df['교육장도로명주소'].apply(myfn2).value_counts()


# 다른 방식
lista = []
for area_str in df['교육장도로명주소'] :
    lista.append(area_str.split()[0])

# print(len(lista))
# print(df['교육장도로명주소'].shape)

s1 = pd.Series(lista)
# s1.unique()
print(s1.value_counts())
```

```python
# 지자체별 온라인/오프라인 강좌 개수

pd.set_option('display.max_columns',None) # 전체 열 출력하기
pd.set_option('display.max_rows',None) # 전체 행 출력하기


df1 = pd.concat((s1, df['교육방법구분']), axis=1, ignore_index=True)
df1.columns = ['교육장도로명주소', '교육방법구분']
# print(df1)

group_df = df1['교육방법구분'].groupby(df1['교육장도로명주소'])

# print(group_df.value_counts().sort_values(ascending=False))

for i in df1['교육장도로명주소'].unique() :
  print(f'{i} {group_df.value_counts().sort_values(ascending=False)[i]}')
```

```python
for i in df1['교육장도로명주소'].unique() :
  filter1 = df1['교육장도로명주소'] == i
  print(f'--------{i}--------')
  for j in df1['교육방법구분'].unique() :
    filter2 = df1['교육방법구분'] == j
    print(j, df1.loc[filter1 & filter2, '교육방법구분'].count())
  print('\n')
    # df.loc[filter1, '교육방법구분'].count()
```
