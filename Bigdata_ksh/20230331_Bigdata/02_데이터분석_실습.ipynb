### data.go.kr : 전국도서관표준데이터

- 전국 단위 도서관 유형별 개수 : groupby, count 활용
- 시도별 도서관 개수 : groupby, count 활용
- 시도별 도서관 유형별 개수 : groupby, count 활용
- 시도별 도서관 유형별 자료수(도서) 평균, 합 : groupby, agg 활용

- 전국 단위 : 열람좌석수 최대 도서관 : describe, max 활용
- 전국 단위 : 자료수 최대, 최소 도서관 : max, min 활용

```python
pd_data = pd.read_csv('전국도서관표준데이터.csv', encoding='cp949')

# print(pd_data.head())

for i,n in enumerate(pd_data.columns):
    print(i, n)
```

```python
sub_data = pd_data[['시도명', '도서관유형', '자료수(도서)', '열람좌석수']]
```

```python
# 전국 단위 도서관 유형별 개수
g1 = sub_data.groupby(['도서관유형'])
g1.count()
```

```python
g4 = sub_data.groupby(['시도명'])
g4.count()
```

```python
# 시도별 도서관 유형별 개수
g2 = sub_data.groupby(['시도명', '도서관유형'])
g2.count()
```

```python
# 시도별 도서관 유형별 자료수(도서) 평균, 합
g3 = sub_data.groupby(['시도명', '도서관유형'])
df_g3 = g3.agg([np.mean, np.sum])
df_g3['자료수(도서)']
```

```python
# 전국 단위 : 열람좌석수 최대 도서관
sub_data['열람좌석수'].describe()
sub_data['열람좌석수'].max()
```

```python
# 전국 단위 : 자료수 최대, 최소 도서관
sub_data['자료수(도서)'].describe()
print('최대 :',sub_data['자료수(도서)'].max(), '최소 :', sub_data['자료수(도서)'].min())
```

```python
# by filter

filter1 = sub_data['열람좌석수'] == sub_data['열람좌석수'].max()
filter2 = sub_data['열람좌석수'] == sub_data['열람좌석수'].min()
filter3 = sub_data['자료수(도서)'] == sub_data['자료수(도서)'].max()
filter4 = sub_data['자료수(도서)'] == sub_data['자료수(도서)'].min()

sub_data[filter4]
```

### kosis.kr : 2021년 시도별 인구

- 취합하여 하나의 df로 완성(pd.concat 활용)
- 시도명(인덱스), 도서관 개수, 인구, 인구1인당 도서관 수

```python
population_data = pd.read_csv('2021_population.csv', encoding='cp949')

population_data = population_data.iloc[1:,:]

population_data['행정구역별(읍면동)'].unique()
sub_data['시도명'].unique()

population_data.set_index('행정구역별(읍면동)', inplace=True)
population_data.rename(columns={'2021':'인구'}, inplace=True)
population_data
```

```python
print(len(sub_data['시도명'].unique()))
print(len(population_data))
```

```python
g_df = sub_data.groupby('시도명').count()
g1_df = g_df[['도서관유형']]
g1_df.rename(columns= {'도서관유형': '도서관개수'}, inplace=True)
g1_df
```

```python
df = pd.concat([g1_df, population_data], axis=1)
df
```

```python
lib_per = df['도서관개수'].astype('float64')/df['인구'].astype('float64')

df = pd.concat([df, lib_per], axis=1)

df.rename(columns={0:'인구1인당 도서관 수'}, inplace=True)

df
```
