### kosis.kr : 2021년 시도별 인구

- 취합하여 하나의 df로 완성(pd.concat 활용)
- 시도명(인덱스), 도서관 개수, 인구, 인구1인당 도서관 수

```python
pd_2021_population = pd.read_csv('2021_population.csv', encoding='cp949', header=1)

# header=1 을 적용하지 않는 경우 drop으로 조치
# pd_2021_population.drop(0, axis=0, inplace=True)  # copy본을 만들어냄

# pd_2021_population = pd_2021_population.iloc[1:,:]

pd_2021_population    # 데이터가 많은 경우 .head() 활용 권장
```

```python
lib_count = sub_data.groupby('시도명').count()

# pd_2021_population.set_index('행정구역별(읍면동)', inplace=True)
pd_2021_population = pd_2021_population.set_index('행정구역별(읍면동)')

print(pd_2021_population)

print(lib_count)
```

```python
df10 = pd.concat((pd_2021_population, lib_count), axis=1)

df10['ratio'] = df10['총인구 (명)'].astype(np.float64)/df10['도서관유형'].astype(np.float64)
df10.insert(1, 'ratio2', df10['도서관유형'].astype(np.float64)/df10['총인구 (명)'].astype(np.float64))
```
