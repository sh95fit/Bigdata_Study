```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd_data = pd.read_csv('전국도서관표준데이터.csv', encoding='cp949')

# print(pd_data.head())

for i,n in enumerate(pd_data.columns):
    print(i, n)
```

```python
sub_df = pd_data[['시도명', '도서관유형', '자료수(도서)']]

print(sub_df)
```

```python
g1 = sub_df.groupby(['시도명', '도서관유형'])
df1 = g1.count()
df1.index = df1.index.swaplevel(0,1)
df1
```

```python
df2 = g1.agg([np.sum, np.max])
# df2.columns = df2.columns.swaplevel(0,1)
df2
```

```python
df2 = df2.stack()
df2
```

```python
df3 = df2.unstack()
df3
```

```python
df4 = df3.unstack()
df4
```
