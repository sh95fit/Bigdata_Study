```python
def check_df(df) :
    # print(df.head())
    for i, col_name in enumerate(df) :
        print(i, col_name, df[col_name].dtype)
        if df[col_name].hasnans :
            print(df[col_name].isna().value_counts())
        print('')
        if len(df[col_name].unique()) < 10 :
            print(df[col_name].value_counts())
        print('')
```

```python
pd_data = pd.read_csv('전국자동차정비업체표준데이터.csv', encoding='cp949')

pd_data.head()

for i,n in enumerate(pd_data.columns):
    print(i, n)
```

```python
check_df(pd_data)
```

```python
def address_list(full_address) :
  return full_address.split()[0]
```

```python
def rewrite(area_name) :
    if area_name == '용인시' :
        return '경기도'
    elif area_name == '동대문구' :
        return '서울특별시'
    elif area_name == '고창군청' :
        return '예외'
    elif area_name == '계룡시청' :
        return '대전광역시'
    elif area_name == '서울특별시중구청' :
        return '서울특별시'
    elif area_name == '제주특별자치도청' :
        return '제주특별자치도'
    elif area_name == '괴산군청' :
        return '예외'
    elif area_name == '군위군청' :
        return '예외'
    elif area_name == '함안군' :
        return '예외'
    elif area_name == '서대문구청' :
        return '서울특별시'
    elif area_name == '은평구청' :
        return '예외'
    else :
        return area_name
```

```python
# 지자체별(광역시도) 자동차 정비업체 개수
# pd_data['소재지도로명주소'].unique()
pd_data['관리기관명'].unique()
```

```python
addr_df = pd_data['관리기관명'].apply(address_list)
addr_df = addr_df.apply(rewrite)
addr_df.unique()
```

```python
len(addr_df.unique())
```

```python
sub_data = pd_data[['관리기관명','영업상태', '데이터기준일자']]
sub_data
```

```python
sub_data['관리기관명'] = addr_df
sub_data
```

```python
sub_data_1 = pd_data[['관리기관명','영업상태']]
sub_data_1['관리기관명'] = addr_df
df1 = sub_data_1.groupby(['관리기관명'])
df1 = df1.count()
df1.rename(columns={'영업상태':'자동차정비소개수'}, inplace=True)
df1
```

```python
df = sub_data.groupby(['관리기관명', '영업상태'])
g1 = df.count()
g1
```

```python
df = g1.unstack()
df.rename(columns={1:'영업',2:'휴업',3:'폐업'}, inplace=True)
df.rename(columns={'데이터기준일자':'영업상태'}, inplace=True)
# df
df = pd.concat([df,df1], axis=1)
df.fillna(0, inplace=True)
df

```

```python
test_data = pd_data[['소재지도로명주소', '소재지지번주소', '영업상태', '데이터기준일자']]

filter1 = test_data['소재지도로명주소'].isna()

test_data['소재지지번주소'][filter1]

test_data['소재지도로명주소'][filter1] = test_data['소재지지번주소'][filter1]

print(test_data.isna().sum())

sub_data = test_data[['소재지도로명주소', '영업상태']]
sub_data

sub_data['소재지도로명주소']= sub_data['소재지도로명주소'].apply(address_list)
g3 = sub_data.groupby('소재지도로명주소')
df3 = g3.count()
df3
```

```python
sub_data_2 = test_data[['소재지도로명주소', '영업상태', '데이터기준일자']]
sub_data_2['소재지도로명주소']= sub_data['소재지도로명주소'].apply(address_list)
df = sub_data_2.groupby(['소재지도로명주소', '영업상태'])

df = df.count()
df = df.unstack()
df.rename(columns={'데이터기준일자':'영업상태', 1:'영업',2:'휴업',3:'폐업'}, inplace=True)
df = pd.concat([df, df3], axis=1)
df.fillna(0, inplace=True)

df
```
