```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
xs = np.linspace(0, 50, 100)
ys = np.sin(xs)
print(xs)
print(ys)
```

```python
_, axe = plt.subplots()

axe.plot(xs, ys)
```

```python
fig, axe = plt.subplots()

axe.plot(xs, ys)

fig.set_facecolor('red')
fig.set_edgecolor('white')
fig.suptitle('Figure_1')
fig.supxlabel('Figure1_X')
fig.supylabel('Figure1_Y')
fig.text(0.5, 0.7, 'mytext')

axe.set_facecolor('yellow')

axe.set_title('axe_title')
axe.set_xlabel('axe_x')
axe.set_ylabel('axe_y')
axe.set_xlim(0, 50)
axe.set_ylim(-2, 2)
axe.set_xticks(np.linspace(0, 50, 5))
axe.set_xticklabels(list('abcde'))
axe.set(facecolor='blue', title='axe_title_2')
```

```python
fig, axes = plt.subplots(2,3, figsize=(12,4))    # figsize(, )  :   / default : 6.4, 4.8

print(axes)
# axes.plot(xs,ys)

axes[1][1].plot(xs, ys)
axes[0][0].plot(xs, ys)
```
