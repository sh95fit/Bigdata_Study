```python
pd_data = pd.read_csv('연도별_전력수급_실적_20230405210501_long.csv', encoding='cp949')

pd_data.head()
```

```python
fig1 = plt.figure()

axe1 = fig1.add_subplot(1,1,1)

col_names = list(pd_data.columns)
col_names[0] = '년도'
pd_data.columns = col_names


x = pd_data['년도']
y = pd_data['경제성장률 (%)']

axe1.plot(x,y)

fig1.suptitle('년도별 경제성장률')

axe1.set_xlabel('년도')
axe1.set_ylabel('경제성장률')
```
