```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import matplotlib

matplotlib.rcParams

matplotlib.rcParams['font.family']

current_font_list = matplotlib.rcParams['font.family']

font_path = 'C:\\Windows\\Fonts\\batang.ttc'

kfont = matplotlib.font_manager.FontProperties(fname=font_path).get_name()

matplotlib.rcParams['font.family'] = [kfont] + current_font_list

current_font_list
```

```python
pd_data = pd.read_csv('자동차등록대수현황_시도별_20230410201013.csv', encoding='cp949', header=[1,2])
# pd_data.drop(('시도명(1)', '시도명(1)'), axis=1, inplace=True)    # 멀티인덱스는 튜플로 삭제해줘야함
pd_data.drop('시도명(1)', axis=1, inplace=True) 
pd_data.head()
```

```python
pd_data.set_index(('시군구(1)','시군구(1)'), inplace=True)
# pd_data.set_index('시군구(1)', inplace=True)

pd_data.index.name='지역구'

pd_data
```

```python
fig = plt.figure(figsize=(12,8), layout='constrained')

axe1 = fig.add_axes(221)
axe2 = fig.add_axes(222)
axe3 = fig.add_axes(223)
axe4 = fig.add_axes(224)

# plt.subplots_adjust(left=0.125,
#                     bottom=0.1, 
#                     right=0.9, 
#                     top=0.9, 
#                     wspace=0.2, 
#                     hspace=0.35)


x = pd_data[('승용', '관용')]
y = pd_data[('화물', '관용')]

fig.suptitle('부산 자동차 등록대수 현황')

axe1.bar(x.index, x, color=('r','b'), label=x.index)
axe1.set_title('지역별 승용-관용 자동차 대수 bar 차트')
axe1.set_xticklabels(x.index, rotation=45)
axe1.legend(['연제구', '해운대구', '영도구'], loc='center', ncols=2)

colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(x)))
ex_array = np.zeros(len(pd_data.index))
ex_array[-3] = 0.15
axe2.pie(x, labels=pd_data.index, autopct='%1.1f%%', colors=colors, radius=1, center=(4, 4), explode=ex_array)



axe3.scatter(x, y, color='red', marker='*')
axe3.set_title('승용-관용 vs. 화물-관용 산점도')

axe4.plot(x.index, x, 'bo--', label='승용-관용')
axe4.set_title('지역별 승용-관용 자동차 대수 line 차트')
axe4.legend()
axe4.set_xticklabels(x.index, rotation=45)
```

```python
fig, axes = plt.subplots(2,2, figsize=(12,8), layout='constrained') # 'constrained' 'compressed', 'tight', 'none', LayoutEngine, None

# plt.subplots_adjust(left=0.125,
#                     bottom=0.1, 
#                     right=0.9, 
#                     top=0.9, 
#                     wspace=0.2, 
#                     hspace=0.35)


axes[0,0].bar(pd_data.index, pd_data[('승용','관용')], color=['y','m'], label=pd_data.index)
axes[0,0].set_xticklabels(pd_data.index, rotation=45)
axes[0,0].legend(['연제구','영도구','해운대구'], loc='center', ncols=2)
axes[0,0].set_title('승용-관용 bar Chart')


ex_array = np.zeros(len(pd_data.index))
ex_array[-3] = 0.1
axes[0,1].pie(pd_data[('승용','관용')], labels=pd_data.index, autopct='%1.1f%%', explode=ex_array)  # '%' 이스케이프 문자! (역슬래시와 유사한 기능!)
axes[0,1].set_title('승용-관용 pie Chart')

axes[1,0].scatter(pd_data[('승용','관용')], pd_data[('화물','관용')], s=20, color='r', marker='*')
axes[1,0].set_title('승용-관용 vs. 화물-관용 산점도')

axes[1,1].plot(pd_data.index, pd_data[('승용','관용')], 'bo--', label='승용-관용')
axes[1,1].set_xticklabels(pd_data.index, rotation=45)
axes[1,1].set_title('승용-관용 line Chart')
axes[1,1].legend()
```
