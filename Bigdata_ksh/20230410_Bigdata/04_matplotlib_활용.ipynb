```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import matplotlib

matplotlib.rcParams

matplotlib.rcParams['font.family']

current_font_list = matplotlib.rcParams['font.family']

font_path = 'C:\\Windows\\Fonts\\batang.ttc'

kfont = matplotlib.font_manager.FontProperties(fname=font_path).get_name()

matplotlib.rcParams['font.family'] = [kfont] + current_font_list

# current_font_list
```

```python
pd_data = pd.read_csv('자동차등록대수현황_시도별_20230410201013.csv', encoding='cp949', header=[1,2])
# pd_data.drop(('시도명(1)', '시도명(1)'), axis=1, inplace=True)    # 멀티인덱스는 튜플로 삭제해줘야함
pd_data.drop('시도명(1)', axis=1, inplace=True) 
pd_data.head()
```

```python
pd_data.set_index(('시군구(1)','시군구(1)'), inplace=True)

pd_data.index.name = '지역구'

pd_data
```

```python
pd_data = pd_data.stack()
```

```python
pd_data.index.name = ('지역구', '차종')

pd_data.index
```

```python
pd_data.reset_index(inplace=True)

pd_data.rename(columns={'level_1':'차종'}, inplace=True)

pd_data
```

```python
g1 = pd_data.groupby(['지역구','차종'])

df1 = g1.sum()

data_1 = df1.loc[('북구','영업용')]
data_2 = df1.loc[('북구','관용')]
data_3 = df1.loc[('북구','자가용')]
```

```python
fig = plt.figure(figsize=(12,8))

plt.subplots_adjust(left=0.125,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.2, 
                    hspace=0.35)

axe1 = fig.add_axes(211)
axe1.bar(data_1.index, data_1, color=['r', 'g'])
axe1.set_title('북구 영업용 - 차종별')

axe2 = fig.add_axes(223)
axe2.pie(data_2, labels=data_2.index, autopct='%1.2f%%')
axe2.set_title('북구 관용 - 차종별')

axe3 = fig.add_axes(224)
axe3.pie(data_3, labels=data_3.index, autopct='%1.0f%%')
axe3.set_title('북구 자가용 - 차종별')
```
