```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
pd_iris = pd.read_csv(url, encoding='cp949', header=None)

pd_iris.head()
```

```python
fig, axe = plt.subplots()

axe.hist(pd_iris[0], bins=20, edgecolor='w', linewidth=0.5)    # 분포, x값 : continuous variable
# 히스토그램의 경우 구간을 나눠 구간별로 분포를 나타내는 경우 활용
# bar는 값별로 얼마가 나왔는지 나타냄

# axe.plot(pd_iris[0])
# axe.scatter(np.arange(len(pd_iris[0])), pd_iris[0])
```

```python
s1 = pd_iris[0].value_counts()
s1 = s1.sort_index()

fig, axe = plt.subplots()

axe.bar(s1.index, s1)   # x값 : categorical variable
```

```python
fig, axe = plt.subplots()

axe.boxplot([pd_iris[0],pd_iris[1]]) # 최상단 : max 값, 최하단 : min 값, 박스 상단 : 75% 값, 박스 하단 : 25% 값, 주황선 : mean
pd_iris[0].describe()
```
