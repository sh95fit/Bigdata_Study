```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd_tips = sns.load_dataset('tips')

pd_tips.head()
```

```python
# sns.relplot(data=pd_tips, x='total_bill', y='tip')
# sns.relplot(data=pd_tips, x='day', y='tip')   # 상관 없는 데이터

# seaborn에서는 hue, style, size 활용 가능!  <- categorical variable
# hue : 색상으로 카테고리 구분
# style : marker 모양으로 카테고리 구분
# size : marker 사이즈로 카테고리 구분
sns.relplot(data=pd_tips, x='total_bill', y='tip', hue='sex', style='smoker', size="size")
```

```python
# col은 카테고리별 그래프를 좌우로 구분
# row는 카테코리별 그래프를 위아래로 구분
sns.relplot(data=pd_tips, x='total_bill', y='tip', col='sex', row='smoker')
```

```python
sns.relplot(data=pd_tips, x='total_bill', y='tip', kind='line')  # default kind : scatter

# kind를 통해 차트 형태를 바꿔줄 수 있다!
```

```python
x = sns.relplot(data=pd_tips, x='total_bill', y='tip')  # figure level
print(type(x), type(x.figure))

fig = x.figure
fig.suptitle('my relplot')
print(type(x.axes))     #ndarray
print(x.axes, type(x.axes[0,0]))

axe = x.axes[0,0]
axe.set_xticklabels(list('abcde'), rotation=45)
```

```python
fig, axes = plt.subplots(2,2)

sns.scatterplot(data=pd_tips, x='total_bill', y='tip', ax=axes[0,1])    # axe level
x = sns.lineplot(data=pd_tips, x='total_bill', y='tip', ax=axes[1,0])    # axe level

print(type(x))
```
