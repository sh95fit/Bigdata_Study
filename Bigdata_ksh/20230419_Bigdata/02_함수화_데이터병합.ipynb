```python
def criminal_drug(filename, drop, col_order, division) :
  pd_data = pd.read_csv(filename, encoding='cp949', header=[0,1,2])

  for i in drop :
    pd_data = pd_data.drop((i), axis=1)

  pd_data = pd_data.stack(level=0)

  pd_data.columns = col_order

  pd_data = pd_data.replace('-', '0')

  pd_data['마약'] = pd_data['마약'].astype(np.int64) + pd_data['마약류'].astype(np.int64)

  pd_data = pd_data.drop('마약류', axis=1)

  pd_data = pd_data[['마약', '대마', '항정신성의약품', '환각물질', '알코올', '미상', '해당무', '합계']]

  pd_data = pd_data.fillna(0)

  pd_data = pd_data.reset_index()

  pd_data = pd_data.set_index('level_1')
  pd_data = pd_data.drop('level_0', axis=1)
  pd_data.index.name = '시점'

  pd_data['구분'] = pd.Series([division]*len(pd_data), index=pd_data.index)

  pd_data = pd_data.reset_index()

  pd_data = pd_data.set_index(['구분', '시점'])

  return pd_data
```
